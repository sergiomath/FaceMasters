```python
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_lfw_people
```

```python
faces =fetch_lfw_people(min_faces_per_person=60) ##se descargan imagenens al azar de figuras p√∫blicas mientras no se tenga una base propia de reconocimiento 
```

```python
print(faces.target_names)
```

```python
print(faces.images.shape)## 62px de altura , 47px de ancho
```

```python
fig,ax=plt.subplots(5,5,figsize=(16,9))
for i,ax_i in enumerate(ax.flat):
    ax_i.imshow(faces.images[i],cmap="bone")
    ax_i.set(xticks=[],yticks=[],xlabel=faces.target_names[faces.target[i]])
```

```python
62*47 ##n√∫mero completo de pixeles 
```

```python
from sklearn.svm import SVC ##Suport vector machine clasifier, 
from sklearn.decomposition import PCA ## An√°lisis de componentes principales para reducir dimensionalidad
from sklearn.pipeline import make_pipeline
```

```python
pca=PCA(n_components=150,whiten=True,random_state=42)
svc=SVC(kernel="rbf",class_weight="balanced") ##rbf clasificar por rasgos circulares , clasificador pondera los rasgos mas importantes (balanced)
model =make_pipeline(pca,svc)
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(faces.data,faces.target,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
    "svc_C":[0.1,1,5,10,50],
    "svc_gamma":[0.0001,0.0005,0.001,0.005,0.01]
}
grid=GridSearchCV(model,param_grid)

grid.fit(Xtrain,Ytrain)
```
