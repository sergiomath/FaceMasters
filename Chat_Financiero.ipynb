```python
from neuralintents import GenericAssistant
import matplotlib.pyplot as plt
import pandas as pd
import pandas_datareader as web
import mplfinance as mpf
import pickle
import sys
import datetime as dt
import pyttsx3 as tts
from pandas_datareader import data as pdr
import fix_yahoo_finance as yf 
```

```python
portafolio={'AAPL':20,"TSLA":5,"GS":5}

with open('portafolio.pkl','wb') as f:
    pickle.dump(portafolio,f)
    
with open('portafolio.pkl','rb') as f:
    portafolio = pickle.load(f)
```

```python
def save_portafolio():
    with open('portafolio.pkl','wb') as f:
        pickle.dump(portafolio,f)
```

```python
def add_portafolio():
    ticker = input("Que accion quiere agregar: ")
    amount = input("cuantas acciones quiere agregar: ")
    if ticker in portafolio.keys():
        portafolio[ticker] += int(amount)
    else:
        portafolio.append(ticker)
        portafolio[ticker] + int(amount)
    save_portafolio()
```

```python
def remove_portafolio():
    ticker = input("Cual accion quiere vender: ")
    amount = input("Cuantas acciones quiere vender: ")
    if ticker in portafolio.keys():
        if amount <= portafolio[ticker]:
            portafolio[ticker] -= amount
            save_portafolio()
        else:
            print('No tiene suficientes acciones')
    else:
        print(f"No se tiene acciones de {ticker}")
```

```python
def show_portafolio():
    print("tu portafolio es: ")
    for ticker in portafolio.keys():
        print(f"Posee {portafolio[ticker]} acciones de {ticker} ")
```

```python
def portafolio_worth():
    sum=0
    for ticker in portafolio.keys():
        data = web.DataReader(ticker,'yahoo')
        price = data['Close'].iloc[-1]
        sum += price
    print(f"El portafolio vale {sum} dolares")
```

```python
def portafolio_gains():
    starting_date = input("Ingrese una fecha para comparar (AAAA-MM-DD): ")
    sum_now = 0
    sum_then = 0
    try:
        for ticker in portafolio.keys():
            data = web.DataReader(ticker,'yahoo')
            price_now = data['Close'].iloc[-1]
            price_then = data.loc[data.index == starting_date]['Close'].values[0]
            sum_now += price_now
            sum_then += price_then
        print(f"Las ganacias relativas: {((sum_now-sum_then)/sum_then)*100}%")
        print(f"Las ganacias absolutas: {sum_now-sum_then} dolares")
    except IndexError:
        print("No se realizaron operaciones en esta fecha")
```

```python
def plot_chart():
    ticker = input("Elija el simbolo de cotizacion: ")
    
    plt.style.use('dark_background')
    
    start = dt.datetime(2020,1,1)
    end = dt.datetime.now()
    
    #data = web.DataReader(ticker,'yahoo',start, end)
    data=pdr.get_data_yahoo(ticker,start=start, end=end)
    #data=yf.download('FB',period = '6mo')
    
    colors = mpf.make_marketcolors(up='#00ff00',down='#ff0000',wick='inherit',edge='inherit',volume='in')
    mpf_style = mpf.make_mpf_style(base_mpf_style='nightclouds',marketcolors=colors)
    mpf.plot(data, type='candle',style=mpf_style,volume=True)
```

```python
def bye():
    print("Hasta luego")
    sys.exit(0)
```

```python
mappings={
    'plot_chart' : plot_chart,
    'add_portafolio' : add_portafolio,
    'remove_portafolio' : remove_portafolio,
    'show_portafolio' : show_portafolio,
    'portafolio_worth' : portafolio_worth,
    'portafolio_gains' : portafolio_gains,
    'salir' : bye
}
```

```python
assistant = GenericAssistant('palabras.json',intent_methods=mappings)
assistant.train_model()
assistant.save_model()

while True:
    message = input("")
    assistant.request(message)
```
